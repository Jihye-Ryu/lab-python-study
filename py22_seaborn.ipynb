# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# tips 데이터 셋

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

# Box plot

연속형 변수의 기술 통계량(최소값, 최대값, 사분위값)

```python
# tip의 box plot
sns.boxplot(data=tips, y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='tip')
plt.show()
```

성별 tip의 box plot

```python
sns.boxplot(data=tips, x='sex', y='tip')
plt.show()
```

```python
tips[tips.sex == 'Female'].tip.describe()
```

요일별 tip의 box plot

```python
sns.boxplot(data=tips, x='day', y='tip')
plt.show()
```

* subplot을 2개 생성(1x2)
* 흡연여부별 tip의 box plot, 시간별 tip의 box plot

```python
fig, ax = plt.subplots(ncols=2, figsize=(8,3))
sns.boxplot(data=tips, x='smoker', y='tip', ax=ax[0])
sns.boxplot(data=tips, x='time', y='tip', ax=ax[1])
plt.show()
```

# Histogram

# Bar plot

## `countplot()`

카테고리 타입 변수의 빈도수를 막대로 표현

```python
sns.countplot(data=tips, x='sex')  # 성별 빈도수 세로 막대 그래프
plt.show()
```

```python
sns.countplot(data=tips, y='sex')  # 성별 빈도수 가로 막대 그래프
plt.show()
```

```python
# 요일 빈도수
sns.countplot(data=tips, x='day')
plt.show()
```

```python
# 요일별 성별 빈도수
tips.groupby(['day', 'sex']).tip.count()
```

```python
sns.countplot(data=tips, x='day', hue='sex')
#> param hue: 카테고리 변수 이름. 카테고리별 다르게 색깔을 사용하는 경우
plt.show()
```

```python
sns.countplot(data=tips, x='sex', hue='day')
plt.show()
```

성별, 흡연 여부별 빈도수

```python
tips.groupby(['sex','smoker']).tip.count()
```

```python
sns.countplot(data=tips, x='sex', hue='smoker')
plt.show()
```

## `barplot()`

* 그룹별 변수의 통계량(평균, 표준편차, 중앙값, 개수, ...)을 막대로 시각화
* 통계량 함수의 기본값은 평균(mean)
* 카테고리 타입 변수와 연속형 변수를 함께 시각화

```python
# 성별 팁의 평균
tips.groupby('sex').tip.mean()
```

```python
sns.barplot(data=tips, x='sex', y='tip')   # estimator 파라미터의 기본값 = np.mean
plt.show()
```

```python
# 성별 tip의 최대값
tips.groupby('sex').tip.max()
```

```python
sns.barplot(data=tips, x='sex', y='tip', estimator=np.max)
plt.show()
```

```python
# 성별, 흡연여부별 팁의 평균
tips.groupby(['sex','smoker']).tip.mean()
```

```python
sns.barplot(data=tips, x='sex', hue='smoker', y='tip')
plt.show()
```

# Scatter plot

* `tip ~ total_bill` 산점도


```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

* `tip ~ total_bill` 산점도
* 점의 색깔을 성별에 따라 다르게 표현

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', alpha=0.8)
plt.show()
```

* `tip ~ total_bill` 산점도
* 점의 색깔을 성별에 따라 다르게 표현
* 점의 모양을 흡연여부에 따라 다르게 표현

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', alpha=0.8, style='smoker')
plt.show()
```

* `tip ~ total_bill` 산점도
* 점의 색깔을 성별에 따라 다르게 표현
* 점의 모양을 흡연여부에 따라 다르게 표현
* 점의 크기를 식사 인원수에 따라 다르게 표현

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', alpha=0.8, style='smoker', size='size')
plt.show()
```

# Category plot

카테고리별로 subplot들을 생성하고 통계값(평균, 중앙값, ...)을 계산해서, 각각의 subplot에 같은 종류의 그래프들을 시각화

```python
# 요일별 팁의 평균 막대 그래프를 흡연여부별로 subplot 생성
sns.catplot(data=tips, x='day', y='tip', col='smoker', kind='bar')  # estimator 기본값 =np.mean
plt.show()
```

```python
sns.barplot(data=tips, x='day', hue='smoker', y='tip')
plt.show()
```

요일별 tip의 평균 막대 그래프 + 시간별 subplot 생성

```python
sns.catplot(data=tips, x='day', y='tip', kind='bar', col='time')
plt.show()
```

요일별 tip의 평균 막대 그래프 + 흡연여부별 subplot + 시간별 subplot 생성

```python
sns.catplot(data=tips, x='day', y='tip', kind='bar', row='smoker', col='time')
plt.show()
```

요일별 성별 tip의 평균 막대 그래프 + 시간별 subplot

```python
sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='bar', col='time')
plt.show()
```

성별 tip의 box plot + 요일별 subplot

```python
sns.catplot(data=tips, x='sex', y='tip', kind='box', col='day')
plt.show()
```
