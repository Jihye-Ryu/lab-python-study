# 파이썬 자료 타입

* 숫자 타입
    * `int`: 정수
    * `float`: 실수

* 문자열 타입
    * `str`

* 논리 타입
    * `bool`: True/False
    * (Java에서는 소문자 true / false. 파이썬은 첫글자 대문자)

* 객체 타입 (클래스 이름 타입?)

파이썬은 변수를 선언할 때 자료 타입을 명시하지 않음! 

변수의 자료 타입은 코드가 실행될 때 결정됨.

```python
# 주석(comment)
# 변수_이름 = 값
n = 123
print(n)
print(type(n))  # type(arg): arg의 자료 타입을 리턴.
```

```python
n = 3.14
print(type(n))
```

식(expression): 값, 변수, 계산식(연산), 함수 호출.

문장(statement): 코드의 최소 실행 단위.

ipynb(IPython notebook) 코드셀 에서 가장 마지막 줄에 있는 문장이 식(expression)이면, 그 식을 평가한 결과를 콘솔 창에 출력해 줌. (print 함수를 사용하지 않아도 됨.)

```python
x = 1
y = 2
# print(x + y)
x + y
```

## 명시적 타입 변환

* `int(값)`: 값을 정수 타입으로 변환해서 리턴.
* `float(값)`: 값을 실수 타입으로 변환해서 리턴.
* `str(값)`: 값을 문자열로 변환해서 리턴.


```python
# colab 노트북의 콘솔에서 입력 받는 방법:
n = input('정수 입력>>> ')
```

```python
type(n)
```

```python
n = int(n)  # 변수 n의 타입을 int 타입으로 변환한 후 n에 저장.
```

```python
type(n)
```

# 산술 연산자

숫자 타입에서 사용되는 산술 연산자: `+, -, *, /, //, %, **`

* `/`: 소수점까지 계산하는 나눗셈.
* `//`: 몫
* `%`: 나머지
* `**`: 거듭 제곱

```python
10 / 3
```

```python
10 // 3
```

```python
10 % 3
```

```python
10 ** 3  # 10 * 10 * 10
```

```python
10 ** -1
```

## 문자열(str) 타입에서 사용할 수 있는 산술 연산자.

* `+`: concatenate. 문자열 이어 붙이기
* `*`: replicate. 문자열 복제

```python
'Hello' + 'Python'
```

```python
'Hello' * 5
```

# 비교 연산자와 논리 연산자

* 비교 연산자: `==, !=, >, >=, <, <=`
* 논리 연산자: `and, or, not`

```python
x = 100
```

```python
(x > 50) and (x < 150)
```

```python
50 < x < 150
```
