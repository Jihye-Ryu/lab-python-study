# Ex 1. 조건문

* 3개의 정수를 입력 받고 각각 변수에 저장
* 3개 정수의 평균을 소수점까지 계산하고 출력
* 평균이 90 이상이면 A, 80 이상이면 B, 70 이상이면 C, 70 미만이면 F를 출력

```python
a = int(input("a :: "))
b = int(input("b :: "))
c = int(input("c :: "))

avg = (a+b+c)/3
print ('평균 점수 = ', avg)

grade = 'A'
if avg >= 90 :
  grade = 'A'
elif avg >= 80 :
  grade = 'B'
elif avg >= 70 :
  grade = 'C'
else :
  grade = 'F'

print ('평점 = ', grade)
```


# Ex 2. 반복문

* 구구단 12단(12x1, ..., 12x12)을 출력


```python
for dan in range(1, 13) :
  print ('========== 구구단 ', dan, '단 ==========\n')
  for i in range(1, dan+1) :
    print (dan, 'x', i, ' = ', dan*i)
  print ('\n')
print ('========= End ========')
```
