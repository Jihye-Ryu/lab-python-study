# `str` 클래스의 메서드

* `str` 클래스의 메서드들은 원본 문자열을 변경하지 않음
* 원본은 유지하고, 원본 문자열을 변경한 새로운 문자열을 생성해 리턴함

```python
s = 'hello, pyTHON! 안녕하세요. 12345 ★☆▷'
```

```python
s.lower()
```

```python
s
```

```python
s.upper()
```

```python
s.capitalize()   # 문자열의 첫글자만 대문자로, 나머지는 소문자로 변경
```

```python
s.count('L')   # 문자열에서 substring이 등장하는 횟수 리턴
```

```python
s.count('L'.lower())
```

```python
url = 'https://www.google.com'
url.startswith('http')
```

```python
url.startswith('ftp')
```

```python
url.endswith('com')
```

```python
phone = '010-0000-0000'
```

```python
phone.split('-')
```

```python
sentence = 'hello python \t 안녕하세요 \nJava'
print(sentence)
```

```python
sentence.split()   # 파라미터 sep=None인 경우, 공백(whitespace: 스페이스, 탭, 줄바꿈)을 기준으로 자름
```

```python
# CSV(comma-seperated values) 파일: 데이터들이 comma(,)로 구분된 형식의 파일
csv = '100,scott,200,scott@oracle.com,1234.56,IT,2022-12-29'
csv.split(',')
```

```python
'/'.join(['2022', '12', '29'])   # 문자열 리스트 -> 합쳐진 문자열
```

```python
s = '               Hello python!                   안녕하세요.......         '
print(s)
print(s.lstrip())    # 문자열의 왼쪽에 있는 공백 제거
print(s.rstrip())    # 문자열의 오른쪽에 있는 공백 제거
print(s.strip())     # 문자열의 양 끝에 있는 공백 제거
```

```python
s = '100::scott::scott@test.com::IT::2022-12-29'
s.replace('::', ',')
```
