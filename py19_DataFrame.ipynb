* 성별에 따른 tip의 차이가 있을까?
    * 남성의 tip의 평균, 여성의 tip의 평균
    * 남성의 tip의 최대값, 여성의 tip의 최대값
    * 남성의 tip의 최소값, 여성의 tip의 최소값
    * 남성의 tip의 중앙값, 여성의 tip의 중앙값

```python
# 성별 컬럼의 unique(중복되지 않는) 값
tips.sex.unique()
```

```python
for s in tips.sex.unique():
    # print(s)
    print(s, 'tip 평균 :', tips[tips.sex == s]['tip'].mean())
```

```python
print('남성 tip 평균: ', tips[tips['sex'] == 'Male'].tip.mean())
print('여성 tip 평균: ', tips[tips['sex'] == 'Female'].tip.mean())
```

```python
print('남성 tip 최대값: ', tips[tips.sex == 'Male'].tip.max())
print('여성 tip 최대값: ', tips[tips.sex == 'Female'].tip.max())
```

```python
print('남성 tip 최소값: ', tips[tips.sex == 'Male'].tip.min())
print('여성 tip 최소값: ', tips[tips.sex == 'Female'].tip.min())
```

```python
print('남성 tip 중앙값: ', tips[tips.sex == 'Male'].tip.median())
print('여성 tip 중앙값: ', tips[tips.sex == 'Female'].tip.median())
```

```python
for s in tips.sex.unique():
    print(s, 'tip 중앙값:', tips[tips.sex == s]['tip'].median())
```

* 흡연 여부별로 팁의 차이가 있을까?
    * 흡연 여부별 팁의 평균
* 요일별로 팁의 차이가 있을까?
    * 요일별 팁의 평균
* 시간별 팁의 차이가 있을까?
    * 시간(Lunch, Dinner)별 팁의 평균

```python
# 흡연 여부별 팁 평균
for sk in tips['smoker'].unique():
    # print(sk)
    print('smoker or non-smoker ', sk, ':', tips[tips.smoker == sk].tip.mean())
```

```python
print('비흡연자 tip 평균:', tips[tips.smoker == 'No'].tip.mean())
print('흡연자 tip 평균:', tips[tips.smoker == 'Yes'].tip.mean())
```

```python
# groupby 이용
tips.groupby(by='smoker')['tip'].mean()    # Series를 리턴
# groupby 괄호 안에 by=  써도 되고 안 써도 됨 (by=' ')는 키워드 방식. 어차피 첫번째 파라미터?가 by= 기 때문에 생략해도 같음.
```

```python
# 요일별 팁 평균
for d in tips['day'].unique():
    print(d, ':', tips[tips.day == d].tip.mean())
```

```python
tips.groupby(by='day').tip.mean()
```

```python
# 시간별 팁 평균
for t in tips.time.unique():
    # print(t)
    print(t, ':', tips[tips.time == t].tip.mean())
```

```python
tips.groupby(by='time')['tip'].mean()
```

```python
print('Dinner 시간대 tip 평균:', tips[tips['time'] == 'Dinner']['tip'].mean())
print('Lunch 시간대 tip 평균:', tips[tips.time == 'Lunch'].tip.mean())
```

# groupby

## 2개 이상의 카테고리별로 groupby

```python
# 성별, 흡연여부별 팁의 평균
tips.groupby(['sex', 'smoker']).tip.mean()
# multi-level index를 갖는 Series 객체를 반환
```

```python
tips.groupby(['smoker', 'sex']).tip.mean()
```

```python
# 성별, 요일별 팁의 평균
tips.groupby(by=['sex', 'day']).tip.mean()
```

```python
tips.groupby(by=['day', 'sex']).tip.mean()
```

## 카테고리별로 2개 이상의 컬럼 선택

```python
# 성별 영수증 금액의 평균, 팁의 평균
tips.groupby('sex')[['total_bill','tip']].mean()    # 데이터프레임을 리턴
```

```python
# 성별, 흡연여부별 영수증 금액과 팁의 평균
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].mean()
```

## 그룹별 2개 이상의 통계 함수 적용

```python
# 성별 팁의 평균, 최대값, 최소값
tips.groupby('sex')['tip'].agg([np.mean, np.max, np.min])   # aggregate
# np.mean()  이런 식으로 괄호 말고 np.mean 이렇게 함수이름만 넘겨야 함!  괄호 붙으면 함수가 아닌, 함수의 계산결과가 넘겨지기 때문
```

```python
# 성별 흡연여부별 영수증 금액과 팁의 평균, 표준편차
# tips.groupby(by=['sex', 'smoker'])[['total_bill','tip']].agg([np.mean, np.std])  # 함수 객체를 넘겨 주거나
tips.groupby(by=['sex', 'smoker'])[['total_bill','tip']].agg(['mean', 'std'])   # 함수 이름 문자열 넘겨줘도 됨
# 데이터프레임이나 Series, 리스트?, 배열?을 전달 받아서 숫자 하나를 리턴하는 함수만 가능!
```
