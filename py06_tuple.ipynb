# tuple(튜플)

* 저장된 원소들을 변경할 수 없는 리스트
* indexing, slicing 가능
* append(), remove() 와 같이, 원소를 변경하는 메서드(기능)는 제공되지 않음

```python
numbers = (1,2,3,10,20,30)
numbers
```

```python
type(numbers)
```

# indexing

인덱스를 사용해 원소 1개를 참조하는 방법

```python
numbers[0]   # 튜플의 첫 번째 원소
```

```python
numbers[-1]  # 튜플의 마지막 원소
```

```python
# 튜플에서 첫 2개 원소 잘라내기
numbers[:2]
```

```python
# 튜플에서 마지막 2개 원소 잘라내기
numbers[-2:]
```

# decomposition(분해)

```python
x,y,z = (1,2,3)
print(f'x = {x}')
print(f'y = {y}')
print(f'z = {z}')
```

1. 2x2 모양의 난수를 저장하는 2차원 리스트 array2 만들기
1. array1과 array2의 같은 인덱스의 원소들끼리 덧셈 결과를 갖는 2차원 리스트 만들기

```python
import random
```

```python
array1 = []
for i in range(2):
    row = [random.randrange(0,10) for _ in range(2)]
    array1.append(row)

for row in array1:
    for x in row:
        print(x, end=' ')
    print()

print(array1)
```

```python
array2 = [
    [random.randrange(0,10) for _ in range(2)] 
    for _ in range(2)
]

for row in array2:
    for x in row:
        print(x, end=' ')
    print()

print(array2)
```
