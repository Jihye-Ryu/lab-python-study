```python
import pandas as pd
import numpy as np
```

```python
file_path = 'https://github.com/JakeOh/202208_itw_java134_lab_python/raw/main/gapminder.tsv'
```

```python
# GitHub에 저장된 gapminder.tsv 파일을 읽고 DataFrame을 생성
gapminder = pd.read_csv(file_path, sep='\t')   # 구분자: 탭(\t)
# CSV(Comma-Seperated Values), TSV(Tab-Seperated Values)
```

```python
# (중복되지 않는) 국가 이름의 개수?
count = 0
for c in gapminder.country.unique():
    count += 1
print(count)
```

```python
# (중복되지 않는) 대륙 이름?
for c in gapminder.continent.unique():
    print(c)
```

```python
# (중복되지 않는) 연도의 개수?
years = gapminder['year'].unique()
print(len(years))
```

```python
# 국가이름, 연도, 1인당 GDP 컬럼을 선택
gapminder[['country', 'year', 'gdpPercap']]
```

```python
# 1인당 GDP의 최댓값, 최솟값을 찾고, 그 row를 출력
# 1인당 GDP 최대값
gapminder[gapminder['gdpPercap'] == gapminder['gdpPercap'].max()]
```

```python
gapminder['gdpPercap'].argmax()
```

```python
gapminder.iloc[[gapminder.gdpPercap.argmax()]]
```

```python
# 1인당 GDP 최소값
gapminder[gapminder['gdpPercap'] == gapminder['gdpPercap'].min()]
```

```python
gapminder.gdpPercap.argmin()
```

```python
gapminder.iloc[[gapminder.gdpPercap.argmin()]]
```

```python
# 기대수명(lifeExp)의 최솟값과 최댓값을 찾고, 그 행들을 출력
# 기대수명 최소값
gapminder[gapminder.lifeExp == gapminder['lifeExp'].min()]
```

```python
# 기대수명 최대값
gapminder[gapminder.lifeExp == gapminder['lifeExp'].max()]
```

```python
gapminder[(gapminder.lifeExp == gapminder.lifeExp.min()) | (gapminder.lifeExp == gapminder.lifeExp.max())]
```

```python
# 일본의 역대 기대수명
gapminder[gapminder.country == 'Japan'][['country', 'year', 'lifeExp']]
```

```python
# 르완다의 역대 기대수명
gapminder[gapminder.country == 'Rwanda'][['country', 'year', 'lifeExp']]
```

```python
# 연도별 기대수명의 평균을 출력
for y in gapminder.year.unique():
    # print(y)
    print(y, '기대수명 평균: ', gapminder[gapminder.year == y]['lifeExp'].mean(), '세')
```

```python
gapminder.groupby('year').lifeExp.mean()
```

```python
# 연도별 1인당 GDP의 평균을 출력
for y in gapminder.year.unique():
    print(y, '1인당 GDP 평균: ', gapminder[gapminder.year == y].gdpPercap.mean())
```

```python
gapminder.groupby('year').gdpPercap.mean()
```

```python
# 연도별 대륙별 전세계 인구를 출력
gapminder.groupby(['year', 'continent'])['pop'].sum()
```

```python
for y in gapminder.year.unique():
    print(y)
    for c in gapminder.continent.unique():
        print(c, gapminder[(gapminder.year == y) & (gapminder.continent == c)]['pop'].sum())
        gapminder[(gapminder.year == y) & (gapminder.continent == c)]['pop']
    print('\n')
```
