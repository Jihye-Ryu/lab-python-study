# set(집합)

* 중복된 데이터를 허용하지 않음: `{1,2,2,3,3,3} = {1,2,3}`
* 데이터 저장 순서가 중요하지 않음: `{1,2,3} = {3,1,2}`
    * 인덱스가 없음 -> indexing, slicing 기능 제공 X

```python
s = {1,2,2,3,3,3}
s
```

[ ] : list

( ) : tuple

{ } : set, dict

set 객체는 list, tuple 객체와 같은 방법으로 for 반복문을 사용할 수 있음

```python
for x in s:
    print(x, end=' ')
```

```python
s.add(10)   # set에 원소 추가
s
```

```python
s.remove(10)   # set에서 원소 삭제
s
```

```python
s1 = {1,2,3,4}
s2 = {2,3,5,6}

print(s1.union(s2))   # 합집합
print(s1.intersection(s2))   # 교집합
print(s1.difference(s2))   # 차집합
print(s2.difference(s1))
```

```python
print(s1 | s2)  # 합집합 연산자
print(s1 & s2)  # 교집합 연산자
print(s1 - s2)  # 차집합 연산자
print(s2 - s1)
```
