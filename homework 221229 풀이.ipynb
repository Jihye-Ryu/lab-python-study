````python
# mapper 메서드 사용해 각 행(1차원 리스트) 원소들을 정수로 변환해준 후, 
# 행 단위로 2차원 리스트에 추가

# score_list =[]
# for row in score_list_str:
#     score_list.append(mapper_to_int(row))     
# score_list
````

````python
# 2차원 리스트 내의 1차원 리스트의 원소들을 하나씩 정수로 변환해 리스트에 저장, 
# 그 리스트들을 새 2차원 리스트로 리턴해 주는 mapper 메서드

def mapper_to_int(iterable):
    result = []
    for row in iterable:
        row_new = []
        for x in row:
            row_new.append(int(x))
        result.append(row_new)
    return result
````

````python
score_list = mapper_to_int(score_list_str)   # 문자열로 이루어졌던 기존 2차원 리스트에 mapper 메서드 적용
score_list
````

````python
score_list = []
for row in score_list_str:
    row_new = [int(x) for x in row]    # list comprehension
    score_list.append(row_new)

score_list
````

````python
def get_total_and_avg(iterable):    # 리스트를 넘겨 주어 합과 평균값을 리턴하는 함수
    total = 0
    for x in iterable:
        total += x
        avg = total / len(iterable)
    return total,avg
````

````python
get_total_and_avg(math_scores)[1]   # 수학 점수들의 평균
# 합과 평균을 리턴하는 함수를 호출해, 인덱스1(평균)만 출력하기
````

````python
def find_max_and_min(iterable):
    max = iterable[0]
    min = iterable[0]
    for x in iterable:
        if x > max:    # 리스트의 원소들을 하나하나씩 비교해서 더 큰 값을 max에 저장
            max = x
        if x < min:    # 리스트의 원소들을 하나하나씩 비교해서 더 작은 값을 min에 저장
            min = x
    return max,min
````

````python
find_max_and_min(math_scores)   # 수학 점수들의 최대값, 최소값 구하기
````
