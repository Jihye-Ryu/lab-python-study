# dict

* dictionary(사전) 형식의 데이터 타입
* 키(key)를 기반으로 값(value)을 저장하는 데이터 타입 (Java의 Map)
    * list, tuple: 인덱스(index)를 기반으로 값을 저장하는 데이터 타입
    * dict에서 키(key)의 역할은 리스트나 튜플에서 인덱스와 비슷한 역할:
    dict의 원소 1개를 참조하기 위한 용도

```python
students = {1: '홍길동', 2: '김길동', 3: '오쌤'}
students
```

```python
type(students)
```

dict에서 키를 사용하는 방법은 리스트나 튜플과 마찬가지로 인덱스 연산자(`[]`)를 사용

```python
students[1]
```

```python
students[10] = 'Scott'   
# 원소 추가: dict[new_key] = new_value
# key는 꼭 연속된 숫자일 필요 X

students
```

```python
students[2] = '허균'   # 원소 변경: dict[old_key] = new_value
students
```

```python
# students.pop(3)   # 원소 삭제
# students

# students.pop(4)    # KeyError (없는 key)
# students 
```

dict에서 key와 value는 다양한 타입의 자료들을 사용할 수 있음

```python
contact = {
    'no': 1,
    'name' : '오쌤',
    'phone' : ['010-0000-0000', '02-000-0000'],
    'email' : {'company':'jake@itwill.co.kr', 'personal':'jake@gmail.com'}
}
contact
```

```python
contact['no']
```

```python
contact['phone']
```

```python
contact['phone'][1]
```

```python
contact['email']['personal']
```

# dict와 for 반복문

for-in 구문에서 dict 타입 객체를 사용하면 key에 대해서 반복(iteration)함

```python
for k in contact:
    print(k)
```

```python
for k in contact:
    print(k, ':', contact[k])
```

```python
contact.items()    # (key, value) 튜플들의 리스트
```

```python
for k,v in contact.items():
    print(k, ':', v)
```

# Dictionary Comprehension

```python
emp_no = [1000, 1001, 2000, 2001]
emp_name = ['King', 'Scott', 'Allen', 'Tiger']
```

```python
emp={}  # emp_no를 키로 하고  emp_name을 값으로 하는 dict
for no, name in zip(emp_no, emp_name):
    emp[no] = name

emp
```

```python
emp2 = {k:v for k,v in zip(emp_no, emp_name)}
emp2
```

```python
messages = ['Merry Chistmasl', 'Happy New Year!', '새해 복 많이 받으세요.' ]

# messages의 문자열을 키로 하고, 그 문자열의 길이를 값으로 하는 dict
message_lengths = {}
for m in messages:
    message_lengths[m] = len(m)

message_lengths
```

```python
{m:len(m) for m in messages}
```
